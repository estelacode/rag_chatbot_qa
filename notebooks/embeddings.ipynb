## Multi-Lingual Models

* embed-multilingual-v3.0

```python
multilingual_embeddings = CohereEmbeddings(
    model="embed-multilingual-v3.0",
    cohere_api_key=os.getenv("COHERE_API_KEY")
)
```

```python
text1 = "Platano"
embeddings1 = multilingual_embeddings.embed_query(text1)
print(embeddings1)
print(len(embeddings1))
```

```python
text2 = "Banana"
embeddings2 = multilingual_embeddings.embed_query(text2)
print(embeddings2)
print(type(embeddings2))
print(len(embeddings2))
```

```python
text3 = "calf"
embeddings3 = multilingual_embeddings.embed_query(text3)

text4 = "cow"
embeddings4 = multilingual_embeddings.embed_query(text4)

text5 = "puppy"
embeddings5 = multilingual_embeddings.embed_query(text5)

text6 = "dog"
embeddings6 = multilingual_embeddings.embed_query(text6)
```

### Visualize Embeddings

```python
data = pd.DataFrame({"text": [text1, text2], 
                     "embeddings": [embeddings1,embeddings2]})
data.head()
```

#### Heatmap

```python
combined_array = np.vstack((data["embeddings"]))
print(combined_array)
```

```python
combined_array.shape
```

```python
type(data["embeddings"][0])
```

```python

# Create a heatmap using matplotlib
vector1 = np.array(data["embeddings"][0]) #(1024,)
vector1 = vector1.reshape(-1,1024)
label1 = data["text"][0]
sns.heatmap(vector1, cmap="Blues", center=0, square=False)
plt.gcf().set_size_inches(20, 2)
plt.title(f'Text:{label1}')
plt.axis('off')
```

```python
# Create a heatmap using matplotlib
vector2 = np.array(data["embeddings"][1]) #(1024,)
vector2 = vector2.reshape(-1,1024)
label2 = data["text"][1]
sns.heatmap(vector2, cmap="Blues", center=0, square=False)
plt.gcf().set_size_inches(20, 2)
plt.title(f'Text:{label2}')
plt.axis('off')
```

```python
print(combined_array.shape)
```

####  PCA

Se aplica PCA para calcular los embeddings en dimension 2D y poder visualizarlos

```python
pca = PCA(n_components=2)
embeddings_pca = pca.fit_transform(combined_array)
```

```python
print(embeddings_pca)
```

```python
data['embeddings_pca'] = embeddings_pca.tolist()
data.head()
```

* visualizar el mapa de embeddings en 2D con altair

```python
df_mapa =pd.DataFrame(data['embeddings_pca'].to_list(),columns=['x','y'])
df_mapa.insert(0,'text',data['text'])
df_mapa.head()
```

```python
alt.Chart(df_mapa, width=600, height=300, title="Embeddings Cohere Multilingual (PCA 2D)").mark_circle(size=60, color="#FF5733").encode(
    x='x',
    y='y',
    tooltip=['text']
)
```

#### UMAP

```python
data2 = pd.DataFrame({"text": [text3, text4, text5, text6], 
                     "embeddings": [embeddings3,embeddings4,embeddings5,embeddings6]})
data2.head()
```

```python
X = np.array(data2["embeddings"].values.tolist())
X.shape
```

```python
import umap

# Assuming you have a dataset called 'X' with high-dimensional data
# Create a UMAP instance with the desired number of components and other parameters
umap_model = umap.UMAP(n_components=2, n_neighbors=2)

# Fit the UMAP model to the data and transform it to the lower-dimensional space
embedddings_umap = umap_model.fit_transform(X)
```

```python
embedddings_umap.shape
```

```python
data2['embedddings_umap'] = embedddings_umap.tolist()
data2.head()
```

```python
df_mapa =pd.DataFrame(data2['embedddings_umap'].to_list(),columns=['x','y'])
df_mapa.insert(0,'text',data2['text'])
df_mapa.head()
```

```python
chart =alt.Chart(df_mapa, width=1000, height=300, title="Embeddings Cohere Multilingual (UMAP 2D)").mark_circle(size=60, color="#FF5733").encode(
    x='x',
    y='y',
)

text = chart.mark_text( align='left',  baseline='middle', dx=7 ).encode(
    text='text'  
)


chart + text  # Combina la capa de puntos con la capa de texto para mostrar los nombres de los puntos en el gráfico
```
